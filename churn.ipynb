Transform Categorical columns into dummy variables 

```python
cats_feat = [1]
```

```python
newdf = pd.get_dummies(df, columns=cats_feat,drop_first=True)
```

```python
newdf
```

```python
newdf1 = newdf.drop([8], axis=1)
newdf1
```

```python
from sklearn.model_selection import train_test_split
X = newdf1.drop([6],axis=1)
y = newdf1[6]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=5)
```

```python
rfc.fit(X_train,y_train)
```

```python
predictions = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import plot_roc_curve
from sklearn.model_selection import train_test_split


X = newdf1.drop([6],axis=1)
y = newdf1[6]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)

svc = RandomForestClassifier(random_state=42)
svc.fit(X_train, y_train)
```

```python
rfc = RandomForestClassifier(n_estimators=10, random_state=42)
rfc.fit(X_train, y_train)
ax = plt.gca()
rfc_disp = plot_roc_curve(rfc, X_test, y_test, ax=ax, alpha=0.9)
#svc_disp.plot(ax=ax, alpha=0.8)
plt.savefig('RF.png', dpi=150)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()

dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)

from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(X_train, y_train)
ax = plt.gca()
dtc_disp = plot_roc_curve(dtc, X_test, y_test, ax=ax, alpha=0.8)
#svc_disp.plot(ax=ax, alpha=0.8)
plt.savefig('DT.png', dpi=150)
plt.show()
```
